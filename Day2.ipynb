```python
# Sample renewable energy sources data
renewable_sources = ["Solar", "Wind", "Hydropower", "Geothermal", "Biomass"]

# Sample green technology project data (for DataFrame)
data = {
    "Project": ["Solar Farm A", "Wind Turbine X", "Hydropower Y", "Solar Roof Z", "Geothermal Plant P"],
    "Technology": ["Solar", "Wind", "Hydropower", "Solar", "Geothermal"],
    "Capacity (MW)": [150, 300, 200, 50, 100],  # Megawatts
    "Cost (Million $)": [200, 400, 350, 100, 250],  # Project cost
    "Location": ["California", "Texas", "Washington", "Nevada", "Idaho"],
    "Completion Year": [2023, 2024, 2022,2025,2023]
}
```

```python
renewable_series = pd.Series(renewable_sources)
print("Renewable energy sources :",renewable_series)
```

```python
df = pd.DataFrame(data)
df
```

```python
#Filter the projects havingcapacity higher than 100 MW
high_capacity_projects = df[df["Capacity (MW)"] > 100]
print("The projects with capacity greater that 100MW :")
high_capacity_projects
```

```python
# Add a new column for cost per MW
df["Cost per MW"] = df["Cost (Million $)"] / df["Capacity (MW)"]
print("Dataframe with column 'Cost per MW': ")
df
```

```python
#Aggegrate the total capacity and cost
total_capacity = df["Capacity (MW)"].sum()
total_cost = df["Cost (Million $)"].sum()
print(f"Total Capacity: {total_capacity} MW")
print(f"Total Cost: {total_cost} Million $")
```

```python
#Grouping solar data
solar_grouped_data = df.groupby("Technology")["Capacity (MW)"].sum()
print("Solar grouped data:")
solar_grouped_data
```

```python
## DATA PREPROCESSING USING NUMPY AND PANDAS
import pandas as pd
import numpy as np
# SAMPLE DATA with missing values
data = {
    "Energy Source": ["Solar", "Wind", "Hydropower", "Geothermal", "Biomass", "Nuclear"],
    "Energy Consumption (MWh)": [1200, np.nan, 2900, np.nan, 2500, 3200],
    "Cost (Million $)": [200, 400, np.nan, 150,250,np.nan]
}
energy_df  = pd.DataFrame(data)
energy_df
```

```python
#count the number of missing values in each column
print("The number of missing values in each column : ")
energy_df.isna().sum()
```

```python
#remove the rows with the missing values
cleaned_df = energy_df.dropna()
print("Dataframe after removing missing values : ")
cleaned_df
```

```python
# Print the second and the third sample from energy_df
print("Second and third sample from energy_df : ")
energy_df.loc[1:3]
```

```python
#Calculating the mean
energy_consumption_mean = energy_df["Energy Consumption (MWh)"].mean()
energy_cost_mean = energy_df['Cost (Million $)'].mean()
#Replacing the NAN values with mean
energy_df['Energy Consumption (MWh)'].fillna(energy_consumption_mean,inplace=True)
energy_df['Cost (Million $)'].fillna(energy_cost_mean,inplace=True)
print("Dataframe after replacing missing values : ")
energy_df
```

```python
#Forward filling the missing values
data = {
    "Energy Source": ["Solar", "Wind", "Hydropower", "Geothermal", "Biomass", "Nuclear"],
    "Energy Consumption (MWh)": [1200, np.nan, 2900, np.nan, 2500, 3200],
    "Cost (Million $)": [200, 400, np.nan, 150,250,np.nan]
}
energy_df  = pd.DataFrame(data)
energy_df.fillna(method='ffill',inplace=True)
print("Dataframe after forward filling missing values : ")
energy_df
```

```python
#Data Scaling using MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
energy_df[["Energy Consumption (MWh)","Cost (Million $)"]] = scaler.fit_transform(energy_df[['Energy Consumption (MWh)','Cost (Million $)']])
print("Data after data scaling with MinMax Scaler : ")
energy_df
```

```python
#Data scaling using StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
energy_df[["Energy Consumption (MWh)","Cost (Million $)"]] = scaler.fit_transform(energy_df[['Energy Consumption (MWh)','Cost (Million $)']])
print("Data after data scaling with Standard Scaler : ")
energy_df
```

```python
#Encoding Categorical Variables
energy_encoded_df = pd.get_dummies(energy_df,columns=['Energy Source'])
print("Data after encoding categorical variables : ")
energy_encoded_df
```

```python
#DATA VISULIZATION
#line plot
import matplotlib.pyplot as plt
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
energy_consumption = [1200,1300,1100,1000,900,800,1700,600,500,400,1300,1200]
plt.plot(months,energy_consumption)
plt.xlabel("Months")
plt.ylabel("Energy Consumption (MWh)")
plt.title("Energy Consumption Over Months")
```

```python
#bar plot
import matplotlib.pyplot as plt
energy_sources = ['Solar','Wind','Hydropower','Geo','Biomass','Nuclear']
energy_consumption = [1200,1300,1100,1000,900,800]
plt.bar(energy_sources,energy_consumption)
plt.xlabel("Energy Sources")
plt.grid(True,linestyle = "--",alpha=0.8)
plt.ylabel("Energy Consumption (MWh)")
plt.title("Energy Consumption by Source")
```

```python
plt.pie(energy_consumption,labels=energy_sources,autopct='%1.1f%%')
plt.title("Energy Consumption by Source")
plt.show()
```

```python
carbon_emission = [12,13,14,15,16,17]
plt.scatter(energy_consumption,carbon_emission)
plt.title("Carbon Emission by Source")
plt.xlabel("Energy Consumption (MWh)")
plt.ylabel("Carbon Emission (kg)")
plt.show()
```
