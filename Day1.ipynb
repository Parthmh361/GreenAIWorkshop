```python
#calculator
inp = input("Enter a operator:")
while(inp != "q"):
    inp = input("Enter a operator:")
    if(inp == 'q'):
        break
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
    if inp == "+":
        print(num1 + num2)
    elif inp == "-":
        print(num1 - num2)
    elif inp == "*":
        print(num1 * num2)
    elif inp == "/":
        if num2 != 0:
            print(num1 / num2)
        else:
            print("Error! Division by zero is not allowed.")
    elif inp == "q":
        break
```

```python
#Check if a year is a leap year or not.
leap_year = lambda year: year % 4 == 0 and (year % 100 != 0 or year %400 ==0)
year = int(input("Enter a year to check if it is a leap year: "))
print(f"is {year} a leap year? {leap_year(year)}")
```

```python
#calculate the averagee carbon_footprint
climate_data = [
    {"city": "City A", "temperature": 25, "carbon_footprint": 500},
    {"city": "City B", "temperature": 30, "carbon_footprint": 350},
    {"city": "City C", "temperature": 22, "carbon_footprint": 600},
    {"city": "City D", "temperature": 15, "carbon_footprint": 200},
    {"city": "City E", "temperature": 28, "carbon_footprint": 450},
]
print("Average carbon footprint: ")
sum = 0
num_of_values =0
for cf in climate_data:
    sum += cf["carbon_footprint"]
    num_of_values +=1
print(sum/num_of_values,"Kg Co2")


```

```python
#filter using "lambda" and "filter" the cities which are sustainable i.e they have a carbon_footprint < 400
sustainable_cities = list(filter(lambda city : city['carbon_footprint'] < 400, climate_data))
print("The sustainable cities are:")
for city in sustainable_cities:
    print(f"{city["city"]}- {city['carbon_footprint']} Kg CO2")
```

```python
#Useing max() to find the city with maximun carbon footprint
max_carbon_footprint_city = max(climate_data, key=lambda city: city['carbon_footprint'])
print("city with the maximum carbon footprint is :")
max_carbon_footprint_city["city"]
```

```python
#Write a function to calculate the carbon footprint
def calc_carbon_footprint(energy_consumption,emission_factor):
    #Calculate the carbon footprint
    carbon_footprint = energy_consumption * emission_factor
    return carbon_footprint
print("Your carbon footprint is:")
calc_carbon_footprint(energy_consumption=100,emission_factor=200)
```
